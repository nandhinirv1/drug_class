```python
#Readin Drug Information created form 1_Readin_process_clinicaltrials_gov_data.ipynb

import pandas as pd
clin_drugs=pd.read_csv('ct_id_drug.txt',delimiter="\t",encoding = "ISO-8859-1",header = None, names = ["nct_id","cleaned_up_drug"] )
print (clin_drugs.head(n=5))

#Preprocess cleaned_up_drug by removing - & space in terms like SGT 53/SGT-53 with none

clin_drugs["processed_drug"]=clin_drugs.apply(lambda x:str(x['cleaned_up_drug']).replace("-","").replace(" ","").lower() if str(x['cleaned_up_drug']).replace(" ","").isalpha()==False else str(x['cleaned_up_drug']).lower() , axis=1)
print (clin_drugs.head(n=5))
```

```python
#Readin crug info class Information created from 2_Readin_chemid_xml.ipynb

import pandas as pd
drug_class=pd.read_csv('chemid_synonym_class_map.txt',delimiter='\t',encoding="ISO-8859-1",names=["chem_id","drug","class_drug_info"],dtype={'chem_id': str, 'drug': str})
print (drug_class.head(n=5))

#Preprocess cleaned_up_drug by removing - & space in terms like SGT 53/SGT-53 with none
drug_class["processed_drug"]=drug_class.apply(lambda x:str(x['drug']).replace("-","").replace(" ","").lower() if str(x['drug']).replace(" ","").isalpha()==False else str(x['drug']).lower() , axis=1)
print (drug_class.head(n=5))


```

```python
#Merge

#dedup fetch only cleaned up drugnames and class_drug_info.
#since same drug could be present in multiple studies, merging on drug  would result in cross product

temp_clin_drugs=clin_drugs[['cleaned_up_drug', 'processed_drug']].drop_duplicates() 
temp_drug_class=drug_class[['processed_drug','class_drug_info']].drop_duplicates()

print (temp_clin_drugs.shape)
print (temp_drug_class.shape)

new=pd.merge(temp_clin_drugs, temp_drug_class, on='processed_drug', how='left')
print (new.shape)
new.sample(100)
```

```python
# Where ever class information is not present, query RxNorm API and fetch class (classtype="ATC1-4","EPC","MOA","VA" in the json output)


import numpy as np

subset_no_class=new[new["class_drug_info"].isnull()]
subset_no_class=subset_no_class[subset_no_class.columns[0:3]]
print(subset_no_class.head(n=5))
print(subset_no_class.shape)
```

```python
remaining_drugs=subset_no_class["cleaned_up_drug"].tolist()
print (len(remaining_drugs))
```

```python
#Do not execute, hits the RxClass API

import requests
import time

#for drug in remaining_drugs: 
base_url = "https://rxnav.nlm.nih.gov/REST/rxclass/class/byDrugName.json?drugName="

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

count_except=0
count=0
with open("remaining_drug_classes.txt","a") as fw:
    fw.write("cleaned_up_drug"+"\t"+"class_umls"+"\n")
    
    for drug in remaining_drugs:
        try:
            count=count+1
            print (count)
            print (drug)
            final_url=base_url+drug
            time.sleep(0.25)
            r = requests.get(url = final_url,headers=headers)
            data=r.json()
            classes=[]
            if "rxclassDrugInfoList" in data.keys():
                for i in range(0,len(data["rxclassDrugInfoList"]["rxclassDrugInfo"])):
                    if data["rxclassDrugInfoList"]["rxclassDrugInfo"][i]["rxclassMinConceptItem"]["classType"] in ["ATC1-4","EPC","MOA","VA"]:
                        classes.append(data["rxclassDrugInfoList"]["rxclassDrugInfo"][i]["rxclassMinConceptItem"]["className"])
                    else:
                        pass
            else:
                pass
            classes=set(classes)
            classes=list(classes)
            #print (classes)
            fw.write(drug+"\t"+"|".join(x for x in classes)+"\n")
        except:
            count_except=count_except+1
        
print ("Exceptions")
print (count_except)

    
```
